```python
%matplotlib inline
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import pandas as pd 

# création du dataset

#importing the Iris dataset with pandas
dataset = pd.read_csv(r'C:\Users\Farouk\Desktop\prediction\qvt_data.csv')
df = pd.DataFrame(dataset)

#n_samples = 100
#random_state = 170
#X, y = make_blobs(n_samples = n_samples, random_state = random_state)
#X = pd.DataFrame(X, columns = ['abscisse','ordonnée'])

# Segmentation à l'aide d'une CAH
y_pred = AgglomerativeClustering(n_clusters = 3).fit_predict(df)

# Visualisation du résultat
plt.figure(figsize = (7,7))
#plt.scatter(X['abscisse'],X['ordonnée'],c = y_pred);
plt.scatter(df['avg_state'], df['user_id'], c = y_pred)
```
